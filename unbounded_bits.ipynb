```python
import pandas as pd
from collections import OrderedDict
fname = "SafeRide_candump_test_secondpart_2018.txt"
mIDs = []
is_unbounded = []
timestemps = []
msgs = []
ids_of_interest = ["1A0", "5A0", "280"]
with open(fname, "r") as f:
    for line in  f:
        # get the message ID and if it not "1A0", "5A0" or "280" don't do noting
        Mid = line[39:42]
        if Mid in ids_of_interest:
            # The file has very specific formating, 
            #so it is possible to parse it only using positional information
            timestemp = line[2:28]
            num_bytes = int(line[46])
            msg = line[50:-1]
            msg_splited = msg.split(" ")
            binary = ""
            for idx in range(num_bytes-1):
                for HEX in msg_splited[idx]:
                    binary_representation = bin(int(HEX, 16))[2:].zfill(4) # convert message to binary 
                    binary += binary_representation
            last_msg_1 = int(msg_splited[num_bytes-1][0], 16)
            last_msg_2 = int(msg_splited[num_bytes-1][1], 16)
            
            # if the last two hex numbers of the message are 0 then the msg. is unbounded
            if last_msg_1 == 0 and last_msg_2 == 0:
                unbounded = False 
            else:
                unbounded = True
            binary += bin(last_msg_1)[2:].zfill(4)
            binary += bin(last_msg_2)[2:].zfill(4)
            mIDs.append(Mid)
            is_unbounded.append(unbounded)
            timestemps.append(timestemp)
            msgs.append(binary)
# create dataframe with the msg. data
df = pd.DataFrame.from_dict(OrderedDict({"mID":mIDs,"Message":binary,"Unbounded":is_unbounded,"Timestemp":timestemps}))
```
